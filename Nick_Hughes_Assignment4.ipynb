```python
import numpy as np  # Import relevant packages
from skimage import io
import matplotlib.pyplot as plt

photo = io.imread('Image_With_Colors.jpeg')
photo.shape # 367 x 612 pixels, 3 values of RGB
plt.imshow(photo) # Shows original image
```

```python
def red_filter(image):
  red_image = np.copy(image) # Make a copy of original image to manipulate
  zero_index = [x for x in range(3) if x != 0] # Creating an index to identify parts of the image that aren't red
  red_image[:,:,zero_index] = 0 # sets non-red parts of image to 0
  return red_image

red_filter(photo)

```

```python
def green_filter(image):
  green_image = np.copy(image)
  zero_index = [x for x in range(3) if x != 1] # same as last function, use a 1 now instead of 0 to identify the non-green parts of image
  green_image[:,:,zero_index] = 0
  return green_image

green_filter(photo)
```

```python
def blue_filter(image):
  blue_image = np.copy(image)
  zero_index = [x for x in range(3) if x != 2]
  blue_image[:,:,zero_index] = 0
  return blue_image

blue_filter(photo)
```

```python
# This code chunk defines new functions for the multicolored filters.
# Now replacing only one x value for 0 versus previously replacing two x values for 0 because filtering for two colors.

def red_and_green_filter(image):
  RG_image = np.copy(image)
  zero_index = [x for x in range(3) if x == 2]
  RG_image[:,:,zero_index] = 0
  return RG_image

def red_and_blue_filter(image):
  RB_image = np.copy(image)
  zero_index = [x for x in range(3) if x == 1]
  RB_image[:,:,zero_index] = 0
  return RB_image

def green_and_blue_filter(image):
  GB_image = np.copy(image)
  zero_index = [x for x in range(3) if x == 0]
  GB_image[:,:,zero_index] = 0
  return GB_image
```

```python
from PIL import Image   # Importing relevant packages
import requests

url = input()
image = np.asarray(Image.open(requests.get(url, stream=True).raw)) # Prompts user to input a url for a jpg image. Url is then converted into a numpy array and assigned the variable "image"

plt.subplot(1,7,1)  # Subplot function used to set images side-by-side. 7 represents number of horizontal subplots, 1 represents that this image will be the first shown of the 7.
plt.imshow(image)
plt.title('Original image') # Title function used to give each image a title

plt.subplot(1,7,2)
plt.imshow(red_filter(image))
plt.title('Red filter')

plt.subplot(1,7,3)
plt.imshow(green_filter(image))
plt.title('Green filter')

plt.subplot(1,7,4)
plt.imshow(blue_filter(image))
plt.title('Blue filter')

plt.subplot (1,7,5)
plt.imshow(red_and_green_filter(image))
plt.title('Red and green filter')

plt.subplot (1,7,6)
plt.imshow(red_and_blue_filter(image))
plt.title('Red and blue filter')

plt.subplot (1,7,7)
plt.imshow(green_and_blue_filter(image))
plt.title('Green and blue filter')

plt.subplots_adjust(left=1, right=5)  # Adjust function used to create more spacing between images, make titles readable.

combined_image = (red_filter(image) + green_filter(image) + blue_filter(image)) # Creating combined image by adding up the separate arrays of color filters individually
print('Is the original image the same as combining the red, green, and blue filtered images?', np.array_equal(combined_image, image)) # Use np.array_equal command to create a boolean to check whether original image array has same numeric values as the combined image array of filters.
```
