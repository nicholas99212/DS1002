```python
#1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

n_trials = 10000
d1 = np.random.randint(1, 7, n_trials)
d2 = np.random.randint(1, 7, n_trials)

R_min = np.minimum(d1, d2)


values, counts = np.unique(R_min, return_counts=True)
probabilities = counts / n_trials

plt.bar(values, probabilities)
plt.xlabel('R_min')
plt.ylabel('Probability')
plt.title('Probability Distribution of R_min')
plt.show()


cumulative_probabilities = np.cumsum(probabilities)

plt.step(values, cumulative_probabilities)
plt.xlabel('R_min')
```

```python
#2
d1 = np.random.randint(1, 7, n_trials)
d2 = np.random.randint(1, 7, n_trials)
d3 = np.random.randint(1, 7, n_trials)

R_med = np.median(np.array([d1, d2, d3]), axis=0)

values, counts = np.unique(R_med, return_counts=True)
probabilities = counts / n_trials

plt.bar(values, probabilities)
plt.xlabel('R_med')
plt.ylabel('Probability')
plt.title('Probability Distribution of R_med')
plt.show()

cumulative_probabilities = np.cumsum(probabilities)

plt.step(values, cumulative_probabilities, where='post')
plt.xlabel('R_med')
plt.ylabel('CDF')
plt.title('CCDF of R_med')
plt.show()
```

```python
#3

results = []
for i in range(n_trials):
    run = True
    sum = 0
    while run is True:
        roll = np.random.randint(1, 7)
        sum += roll
        if roll < 6:
            run = False
            results.append(int(sum))

results = pd.DataFrame(results)

sns.histplot(results)
plt.show()

sns.ecdfplot(results)
plt.show()
```

```python
df = pd.read_csv('/content/mammogram.csv')

# 1.

print( pd.crosstab( df['treatment'], df['breast_cancer_death']))
print('treatment effect:', 44425/44925 - 44405/44910)
# Difference is just .01% between control and mammogram groups

# 2 (Note: I struggled figuring out the code for this part and took much from the answers file)

df['survive'] = 0
df.loc[ df['breast_cancer_death'] == 'no', 'survive'] = 1

df_treat = df.loc[ df['treatment'] == 'mammogram', :]
df_control = df.loc[ df['treatment'] == 'control', :]

S = 5000
fx = []
for t in range(S):
    df_treat_t = df_treat.sample( df_treat.shape[0], axis=0, replace=True)
    df_control_t = df_control.sample( df_control.shape[0], axis=0, replace=True)

    treatment_effect_t = np.mean(df_treat_t['survive']) - np.mean(df_control_t['survive'])
    fx.append(treatment_effect_t)

sns.kdeplot(fx)
plt.show()

sns.ecdfplot(fx)
plt.show()

# 3.
lower = np.quantile(fx, .005)
upper = np.quantile(fx, .995)
print('99% confidence interval:', lower, upper)
# 0 is included in the confidence interval
```
